```python
import pandas as pd
import numpy as np
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import time

df = pd.read_excel('/Users/jaehwan/Desktop/송재환/DS_Project/기술과제 제출/클로젯셰어_과제DB 대체 자료.xlsx', encoding='cp949')
```

```python
df['sub_type'].unique()
```

```python
def price_crwaling(df):
    
    df['brandnm+goodsnm'] = df['brandnm'] + " " + df['goodsnm']
    p_brand = []
    p_goods = []
    p_tag = []
    p_price = []
        
    for i in df['sub_type'].unique():
        
        driver = webdriver.Chrome()
        driver.implicitly_wait(1)
        
        if i == 'BA':
            
            url1 = 'https://closetshare.com/theclozet/front/goods/goods_list_all_page?category=002&sort=1&useType=All'                    

            browser = driver.get(url1)

            for j in np.arange(0,len(df[df['sub_type'] == i]['goodsnm'])):

                    if j % 20 == 0:
                        print(f"{j/len(df[df['sub_type'] == i]['goodsnm'])}")

                    filtered_df = df[df['sub_type'] == i].reset_index()

                    WebDriverWait(driver, 5).until(EC.element_to_be_clickable((By.NAME, 'goodsnm')))
                    driver.find_element_by_name('goodsnm').send_keys(filtered_df['brandnm+goodsnm'][j])
                    driver.implicitly_wait(1)
                    driver.find_element_by_xpath('//*[@id="goodsnmSesrch"]').click()
                    driver.implicitly_wait(1)
                    time.sleep(0.5)

                    if driver.find_element_by_xpath('//*[@id="frmList"]/section/section[3]/div[3]/span').text == 'TOTAL 0':
                        
                        print(driver.find_element_by_xpath('//*[@id="frmList"]/section/section[3]/div[3]/span').text)
                        
                        driver.get(url1)
                        WebDriverWait(driver, 5).until(EC.element_to_be_clickable((By.NAME, 'goodsnm')))
                        driver.find_element_by_name('goodsnm').clear()
                        time.sleep(0.1)

                    elif driver.find_element_by_xpath('//*[@id="frmList"]/section/section[3]/div[3]/span').text != 'TOTAL 0':

                        print(driver.find_element_by_xpath('//*[@id="frmList"]/section/section[3]/div[3]/span').text)
                        
                        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "prod_list_item")))
                        driver.find_element_by_xpath('//*[@id="mainVisual"]/ul/li[2]').click()
                        driver.implicitly_wait(0.5)

                        a = driver.find_element_by_class_name('goodsTopTag').text

                        b = driver.find_element_by_xpath('//*[@id="goodsValue_daily"]').text

                        p_brand.append(filtered_df['brandnm'][j])
                        p_goods.append(filtered_df['goodsnm'][j])
                        p_tag.append(a)
                        p_price.append(b)

                        driver.get(url1)  
                        WebDriverWait(driver, 5).until(EC.element_to_be_clickable((By.NAME, 'goodsnm')))
                        driver.find_element_by_name('goodsnm').clear()
                        time.sleep(0.1)
        
        elif i != 'BA':
            
            url2 = 'https://closetshare.com/theclozet/front/goods/goods_list_all_page?category=001&sort=1&useType=All'

            browser = driver.get(url2)

            for j in np.arange(0,len(df[df['sub_type'] == i]['goodsnm'])):

                    if j % 20 == 0:
                        print(f"{j/len(df[df['sub_type'] == i]['goodsnm'])}")

                    filtered_df = df[df['sub_type'] == i].reset_index()

                    WebDriverWait(driver, 5).until(EC.element_to_be_clickable((By.NAME, 'goodsnm')))
                    driver.find_element_by_name('goodsnm').send_keys(filtered_df['brandnm+goodsnm'][j])
                    driver.implicitly_wait(1)
                    driver.find_element_by_xpath('//*[@id="goodsnmSesrch"]').click()
                    driver.implicitly_wait(1)

                    if driver.find_element_by_xpath('//*[@id="frmList"]/section/section[3]/div[3]/span').text == 'TOTAL 0':

                        driver.get(url2)
                        WebDriverWait(driver, 5).until(EC.element_to_be_clickable((By.NAME, 'goodsnm')))
                        driver.find_element_by_name('goodsnm').clear()
                        time.sleep(0.1)

                    elif driver.find_element_by_xpath('//*[@id="frmList"]/section/section[3]/div[3]/span').text != 'TOTAL 0':

                        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, "prod_list_item")))
                        driver.find_element_by_xpath('//*[@id="mainVisual"]/ul/li[2]').click()
                        driver.implicitly_wait(0.5)

                        a = driver.find_element_by_class_name('goodsTopTag').text

                        b = driver.find_element_by_xpath('//*[@id="goodsValue_daily"]').text

                        p_brand.append(filtered_df['brandnm'][j])
                        p_goods.append(filtered_df['goodsnm'][j])
                        p_tag.append(a)
                        p_price.append(b)

                        driver.get(url2)
                        WebDriverWait(driver, 5).until(EC.element_to_be_clickable((By.NAME, 'goodsnm')))
                        driver.find_element_by_name('goodsnm').clear()
                        time.sleep(0.1)
    
    price_df_list = [p_brand, p_goods, p_tag, p_price]
    price_df = pd.DataFrame(price_df_list).T
    price_df.columns = ['brandnm', 'goodsnm', 'tag', 'price']
    
    return price_df
```

```python
price_df = price_crwaling(df)
```
